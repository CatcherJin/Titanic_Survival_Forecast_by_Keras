```python
import urllib.request
import os

data_url = 'http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.xls'

data_file_path = "data/titanic3.xls"

if not os.path.isfile(data_file_path):
    result = urllib.request.urlretrieve(data_url,data_file_path)
    print("downloaded:",result)
else:
    print(data_file_path,"data file already exists.")
```

```python
#使用Pandas进行数据预处理
import numpy
import pandas as pd

df_data = pd.read_excel(data_file_path)
```

```python
df_data.describe()
```

```python
df_data
```

survival是标签字段 其他是候选特征字段（需要筛选）

```python
#先选取数据
selected_cols = ['survived','name','pclass','sex','age','sibsp','parch','fare','embarked']

selected_df_data = df_data[selected_cols]
```

```python
#先打乱顺序
shuffled_df_data = selected_df_data.sample(frac=1)
```

```python
shuffled_df_data.isnull().any()
```

```python
shuffled_df_data.isnull().sum()
```

```python
#把空值用平均值代替
#sex文字字段用数字字段代替 0 1
#embarked字段有些数据为null可改成某一个值 如：S  且该字段是文字 需要转换为数字

age_mean_value = shuffled_df_data['age'].mean()
shuffled_df_data['age'] = shuffled_df_data['age'].fillna(age_mean_value)

fare_mean_value = shuffled_df_data['fare'].mean()
shuffled_df_data['fare'] = shuffled_df_data['fare'].fillna(fare_mean_value)

shuffled_df_data['embarked'] = shuffled_df_data['embarked'].fillna('S')

shuffled_df_data['sex'] = shuffled_df_data['sex'].map({'female':0,'male':1}).astype(int)

shuffled_df_data['embarked'] = shuffled_df_data['embarked'].map({'C':0,'Q':1,'S':2}).astype(int)
```

```python
shuffled_df_data[:3]
```

```python
#删除name字段  通过drop实现 
#drop不改变原有的df中的数据 而是返回另一个dataframe来存放删除后的数据 axis表示删除列

shuffled_df_data = shuffled_df_data.drop(['name'],axis=1)
```

```python
shuffled_df_data[:3]
```

```python
#分类特征值和标签值

#转换为ndarray数组
ndarray_data = shuffled_df_data.values

#逗号前表示行 逗号后便是列  行都要 列筛选
x_data = ndarray_data[:,1:]#只要第1列到最后（后7列）

y_data = ndarray_data[:,0]#只要第0列
```

```python
x_data[:3]
```

```python
y_data[:3]
```

```python
#特征值需要进行标准化

from sklearn import preprocessing
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))
norm_x_data=minmax_scale.fit_transform(x_data)
```

```python
norm_x_data[:3]
```

```python
train_size = int(len(norm_x_data)*0.8)

x_train = norm_x_data[:train_size]
y_train = y_data[:train_size]

x_test = norm_x_data[train_size:]
y_test = y_data[train_size:]
```

Keras建模

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units=64, #第一层有64个神经元
                               input_dim=7, #输入为7列
                               use_bias=True,
                               kernel_initializer='uniform',
                               bias_initializer='zeros',
                               activation='relu'))
```

```python
model.add(tf.keras.layers.Dropout(rate=0.3))#可以防止过拟合
```

```python
model.add(tf.keras.layers.Dense(units=32,#第2层32个神经元
                               activation='sigmoid'))
```

```python
model.add(tf.keras.layers.Dropout(rate=0.3))
```

```python
model.add(tf.keras.layers.Dense(units=1,#最后一层 一个输出
                               activation='sigmoid'))#因为要输出结果 二分类 用sigmoid好
```

```python
model.summary()
```

```python
#用于模型的保存

logdir = './logs'
checkpoint_path = './checkpoint/Titanic.{epoch:02d}-{val_loss:.2f}.ckpt'

callbacks = [
    tf.keras.callbacks.TensorBoard(log_dir = logdir,
                                  histogram_freq=2),
    tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                       save_weights_only=True,
                                       verbose=1,
                                       period=5)
]
```

```python
#训练模型
model.compile(optimizer=tf.keras.optimizers.Adam(0.003),
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
train_history = model.fit(x=x_train,
                         y=y_train,
                         validation_split=0.2,#取20%的验证集
                         epochs=200,
                         batch_size=40,
                         callbacks=callbacks,
                         verbose=2)
```

```python
train_history.history.keys()
```

```python
import matplotlib.pyplot as plt

def visu_train_history(train_history,train_metric,validation_metric):
    plt.plot(train_history.history[train_metric])
    plt.plot(train_history.history[validation_metric])
    plt.title("Train History")
    plt.xlabel('epoch')
    plt.ylabel(train_metric)
    plt.legend(['train','validation'],loc='upper left')
    plt.show()
```

```python
visu_train_history(train_history,'acc','val_acc')
```

```python
visu_train_history(train_history,'loss','val_loss')
```

评估模型

```python
evaluate_result = model.evaluate(x=x_test,
                                y=y_test)
```

```python
print("loss                 ","acc")
evaluate_result
```
